```python
import matplotlib.pyplot as plt

# lösningsförslag 1 - med listor
t = [0,1,2,3,4,5]
s = [0,1000,1600,2400,3200,4000]

plt.plot(t,s, 'o-')
plt.xlabel("Tiden i timmar")
plt.ylabel("Sträcka i km")
plt.title("S-T diagram för ett flyglan med konstant hastighet")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(6)
y = 800*x
print(y)
print(x)

plt.plot(x,y, 'o-')
plt.xlabel("Tiden i timmar")
plt.ylabel("Sträcka i km")
plt.title("S-T diagram för ett flyglan med konstant hastighet")
plt.show()
```
