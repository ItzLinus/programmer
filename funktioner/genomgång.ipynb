```python
#Definerat en funktion som räknar ut medelvärde av 2 tal. 
def medelvärde(tal1,tal2):
    medel = (tal1+tal2)/2  #lokalvariabel
    return medel

vikt_medel = medelvärde(60,62)

print(f"Min medelvikt är {vikt_medel} k")
```

```python
import numpy as np
import matplotlib.pyplot as matp
def f(x):
    return x**2

#anropa funktionen med parametern 6
print(f(6))

x = np.linspace(-10,10)
y = f(x)
matp.plot(x,y)
matp.xlabel("x")
matp.ylabel("y")
matp.title("f(x) = x²")
matp.grid()
matp.show()
```
